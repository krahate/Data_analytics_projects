```python
def condition(Final_data):
    if (Final_data['Total weight as per Courier Company (Gram)']<500) or (Final_data['Total weight as per Courier Company (Gram)']==500):
        return 0.5
    elif (Final_data['Total weight as per Courier Company (Gram)']>500) and (Final_data['Total weight as per Courier Company (Gram)']<=1000):
        return 1
    elif (Final_data['Total weight as per Courier Company (Gram)']>1000) and (Final_data['Total weight as per Courier Company (Gram)']<=1500):
        return 1.5
    elif (Final_data['Total weight as per Courier Company (Gram)']>1500) and (Final_data['Total weight as per Courier Company (Gram)']<=2000):
        return 2
    elif (Final_data['Total weight as per Courier Company (Gram)']>2000) and (Final_data['Total weight as per Courier Company (Gram)']<=2500):
        return 2.5
    elif (Final_data['Total weight as per Courier Company (Gram)']>2500) and (Final_data['Total weight as per Courier Company (Gram)']<=3000):
        return 3
    elif (Final_data['Total weight as per Courier Company (Gram)']>3000) and (Final_data['Total weight as per Courier Company (Gram)']<=3500):
        return 3.5
    elif (Final_data['Total weight as per Courier Company (Gram)']>3500) and (Final_data['Total weight as per Courier Company (Gram)']<=4000):
        return 4
    else:
        return 4.5
```

```python
Final_data['Weight slab charged by Courier Company (KG)']=Final_data.apply(condition,axis=1)
```

```python
def condn(Final_data):
    if Final_data['Weight slab as per X (KG)']<=0.5:
        if Final_data['Type of Shipment']=='Forward charges':
            if Final_data['Delivery Zone as per X']=='b':
                return(pay_sc['fwd_b_fixed'][0])
            if Final_data['Delivery Zone as per X']=='d':
                return(pay_sc['fwd_d_fixed'][0])
            if Final_data['Delivery Zone as per X']=='e':
                return(pay_sc['fwd_e_fixed'][0])
        if Final_data['Type of Shipment']=='Forward and RTO charges':
            if Final_data['Delivery Zone as per X']=='b':
                return(pay_sc['fwd_b_fixed'][0]+pay_sc['rto_b_fixed'][0])
            if Final_data['Delivery Zone as per X']=='d':
                return(pay_sc['fwd_d_fixed'][0]+pay_sc['rto_d_fixed'][0])
            if Final_data['Delivery Zone as per X']=='e':
                return(pay_sc['fwd_e_fixed'][0]+pay_sc['rto_e_fixed'][0])
    if Final_data['Weight slab as per X (KG)']>0.5:
        if Final_data['Type of Shipment']=='Forward charges':
            if Final_data['Delivery Zone as per X']=='b':
                return(pay_sc['fwd_b_fixed'][0]+ pay_sc['fwd_b_additional'][0])
            if Final_data['Delivery Zone as per X']=='d':
                return(pay_sc['fwd_d_fixed'][0]+ pay_sc['fwd_d_additional'][0])
            if Final_data['Delivery Zone as per X']=='e':
                return(pay_sc['fwd_e_fixed'][0]+pay_sc['fwd_e_additional'][0])
        if Final_data['Type of Shipment']=='Forward and RTO charges':
            if Final_data['Delivery Zone as per X']=='b':
                return(pay_sc['fwd_b_fixed'][0]+pay_sc['rto_b_fixed'][0]+pay_sc['fwd_b_additional'][0]+pay_sc['rto_b_additional'][0])
            if Final_data['Delivery Zone as per X']=='d':
                return(pay_sc['fwd_d_fixed'][0]+pay_sc['rto_d_fixed'][0]+pay_sc['fwd_d_additional'][0]+pay_sc['rto_d_additional'][0])
            if Final_data['Delivery Zone as per X']=='e':
                return(pay_sc['fwd_e_fixed'][0]+pay_sc['rto_e_fixed'][0]+pay_sc['fwd_e_additional'][0]+pay_sc['rto_e_additional'][0])
```

```python
Final_data['Expected Charge as per X (Rs.)']= Final_data.apply(condn,axis=1)
```
